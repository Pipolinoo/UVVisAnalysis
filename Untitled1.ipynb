```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
LEDSignal = pd.read_csv('Baseline References/WhiteLED.txt', delim_whitespace=True, header=None, skiprows=14, names=['wavelength', 'intensity'], dtype={'wavelength':  np.float64, 'intensity': np.float64}, decimal=",")
LEDSignal['intensity_norm'] = LEDSignal['intensity'] / LEDSignal[LEDSignal['wavelength'] > 200]['intensity'].max()

LEDSignal['wavelength'] = LEDSignal['wavelength'] + 12
LEDSignal['intensity_norm_scaled'] = LEDSignal['intensity_norm'] * 1
LEDSignal.plot('wavelength', 'intensity_norm')
```

```python
file = pd.read_csv('./UVVis/InitalProcedure.csv')
reducedFile = file[file.n > 50]
reducedFile = reducedFile[file.n < 135]
spectra = reducedFile.groupby(reducedFile.n)
```

```python
figure2 = plt.figure()

figure2ax1 = figure2.add_subplot(1,1,1)
figure2ax2 = figure2ax1.twinx()
# figure2ax2.set_ylim(-0.5, 6)

figure2ax1.plot(LEDSignal['wavelength'], LEDSignal['intensity_norm_scaled'])


for spectrumIndex in list(spectra.indices.keys()):
    spectrum = spectra.get_group(spectrumIndex).reset_index(drop=True)
    spectrum['intensity'] = spectrum['intensity'].rolling(100).mean()

    spectrumBaseline = spectrum[spectrum.wavelength > 200]
    spectrumBaseline = spectrumBaseline[spectrum.wavelength < 300]
    spectrumBaseline = spectrumBaseline['intensity'].mean()
    spectrum['intensity'] = spectrum['intensity'] - spectrumBaseline
    
    
#     spectrumMaxValue = spectrum[spectrum.wavelength > 590]
#     spectrumMaxValue = spectrumMaxValue[spectrumMaxValue.wavelength < 591]
#     spectrumMaxValue = spectrumMaxValue['intensity'].mean()

    spectrum['intensityNorm'] = spectrum['intensity'] / spectrum['intensity'].max()
    
    
    spectrum['abs'] = -np.log(spectrum['intensity'] / LEDSignal['intensity_norm_scaled'])
    spectrum['absSmoothed'] = spectrum['abs'].rolling(20).mean()
    
    
    figure2ax1.plot(spectrum['wavelength'], spectrum['intensityNorm'])
#     figure2ax2.plot(spectrum['wavelength'], spectrum['absSmoothed'])
```
